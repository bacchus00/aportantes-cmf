```python
# columns of final table
cols = [
    'Archivo',
    'Fecha',
    'Candidato',
    'RUTCandidato',
    'Aportante',
    'RUTAportante',
    'Monto',
    'Eleccion',
    'TerritorioElectoral',
    'Comuna',
    'Region',
    'Distrito',
    'Circunscripcion',
    'CircunscripcionSenatorial',
    'CircunscripcionProvincial',
    'RUTPartido',
    'Partido',
    'Pacto',
    'Subpacto',
    'Afiliacion',
    'TipoDocumento',
    'GlosaDocumento',
    'NumeroDocumento',
    'TipoCuenta',
    'ParaReembolso', 
    'EstadoCuenta',
    'Descripcion1',
    'Descripcion2'
]

raw_data = pd.DataFrame(columns=cols)

# importing files
for i in range(len(file_name_axis)):
  next_df=pd.read_excel(
    '/content/drive/MyDrive/Bushe/Trabajo Dirigido/files/modified/' + file_name_axis[i][0],
    header=file_name_axis[i][1]
  )
  next_df['Archivo'] = file_name_axis[i][0]

  raw_data = pd.concat([raw_data, next_df])
  print('Added:',file_name_axis[i][0])
  print('New Rows:', len(next_df),', Total Rows:', len(raw_data))
```

```python
# funcion to print row value counts
def col_describer(col_name, df):
  print('---'+ col_name +'---')
  print(df[col_name].describe(),'\n')
  print('Nan count:' + str(df[col_name].isna().sum()),'\n')
  print(df[col_name].value_counts(),'\n')
```

```python
# column summaries
for col in raw_data.columns:
  col_describer(col,raw_data)
```
